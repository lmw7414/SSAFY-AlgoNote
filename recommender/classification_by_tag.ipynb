```python
import pandas as pd

from glob import glob

files = glob("./tag/*.csv")


dataframes = []

for file in files:
    df = pd.read_csv(file)

    dataframes.append(df)

result_df = pd.concat(dataframes, ignore_index=True)

print(result_df)
```

```python
grouped_dfs = dict(tuple(result_df.groupby('tag')))

grouped_dfs
```

```python
grouped_dfs = dict(tuple(result_df.groupby('tag')))

# 분할된 데이터프레임을 각각 별도의 변수로 할당하고 CSV 파일로 저장
for tag, group_df in grouped_dfs.items():
    # 'tag' 열 제외
    group_df = group_df.drop('tag', axis=1)
    
    # CSV 파일 이름에 태그 포함
    
    filename = f'./divide/data_{tag}.csv'
    
    # 데이터프레임을 CSV 파일로 저장
    group_df.to_csv(filename, index=False)
    
    print(f"DataFrame for tag '{tag}' saved as '{filename}'")
```
